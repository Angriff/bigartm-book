# Тематическое моделирование с BigARTM

Это — интерактивная книжка по тематическому моделированию с использованием [библиотеки BigARTM](http://bigartm.org/).

Концепции книжки:
- пишем на русском, чтобы уровень входа для новых авторов был достаточно низким
- пишем в первую очередь для себя — чтобы новые студенты группы Воронцова моментально вникали в тему и умели делать эксперименты
- каждый раздел представляет собой законченное, хорошо оформленное исследование
- разделы должны быть независимыми друг от друга, но могут содержать произвольное число тетрадок, скрипты и т.д.
- раздел должен иметь титульную страницу, на которой должно быть:
  - постановки задач
  - мотивация
  - план исследований со ссылками на тетрадки, в которых решаются отдельные задачи
- исследования должны быть воспроизводимы, для этого нужно придерживаться ряда правил
  - приводить ссылки на используемые датасеты
  - приводить ссылки на используемые библиотеки и краткое руководство по их установке
  - код всех операций по приведению данных в необходимый формат писать в тетрадке, либо в скриптах рядом с тетрадкой
  - делать так чтобы перезапуск тетрадок приводил к одному результату
- описание исследований должно быть понятно человеку, не знающему про BigARTM и тематическое моделирование

## Примеры приложений

### [Wikipedia](applications/wikipedia/Main.ipynb)
Эксперименты с многомодальным тематическим моделированием текстов [Википедии](http://www.wikipedia.org/) на разных языках.

### [Хабрахабр](applications/habrahabr/Main.ipynb)
Анализ тематической структуры коллективного блога [Habrahabr.ru](http://habrahabr.ru/).

### [Last.fm](applications/lastfm/Main.ipynb)
Тематическое моделирование плейлистов пользователей интернет-радио [Last.fm](http://www.lastfm.ru/) для автоматического выявления жанров, построения списков похожих артистов и персональных рекомендаций.

### [EUR-Lex](applications/eurlex/Main.ipynb)
Построение классификатора юридических текстов из корпуса [EUR-Lex](http://www.ke.tu-darmstadt.de/resources/eurlex) с большим числом несбалансированных, взаимозависимых, пересекающихся классов. Тематическая модель, обученная с использованием регуляризатора специального вида позволяет строить достаточно точный классификатор при том что обучающая выборка достаточно мала для такого количества тем.
